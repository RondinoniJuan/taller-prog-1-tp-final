```python
from collections import Counter
import matplotlib.pyplot as plt

with open("web-server-log.txt", "r") as file:
    logs = file.readlines()

ips = [log.split()[0] for log in logs]
pages = [log.split()[6] for log in logs]
codes = [log.split()[8] for log in logs]
```

```python
ip_counts = Counter(ips)

ip_labels, ip_values = zip(*ip_counts.items())
plt.figure(figsize=(10, 6))
plt.bar(ip_labels, ip_values, color='skyblue')
plt.xticks(rotation=90)
plt.title("Cantidad de accesos por IP")
plt.xlabel("Dirección IP")
plt.ylabel("Número de accesos")
plt.tight_layout()
plt.show()
```

```python
status_counts = Counter(codes)

labels, sizes = zip(*status_counts.items())
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140, colors=['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3'])
plt.title("Proporción de códigos HTTP")
plt.show()
```

Gráfico de Barras: Este gráfico muestra la cantidad de accesos al servidor por cada dirección IP única. Permite identificar cuáles IPs accedieron más frecuentemente al servidor.

Gráfico Circular: Este gráfico muestra la proporción de diferentes códigos HTTP en los accesos registrados. Ayuda a visualizar el porcentaje de accesos exitosos (código 200) frente a otros códigos de estado.
